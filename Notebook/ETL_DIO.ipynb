Criando DataFrame

```python
import pandas as pd

# Criando um conjunto de dados com vários usuários
data = {
    'UserID': [1, 2, 3, 4, 5],
    'Name': ['Alice Silva', 'Bruno Souza', 'Carla Dias', 'Diego Ramos', 'Elena Luz']
}

# Salvando como CSV
df_input = pd.DataFrame(data)
df_input.to_csv('Data.csv', index=False)
print("Arquivo Data.csv criado com sucesso!")
```

Extract

```python
# Extraindo os IDs e Nomes do CSV
df = pd.read_csv('Data.csv')
users = df.to_dict(orient='records')

# Preparando a estrutura 'news' para cada usuário
for user in users:
    user['news'] = []

print(f"Extraídos {len(users)} usuários do arquivo CSV.")
```

Transform

```python
import os
from google.colab import userdata
from groq import Groq

# Utiliza a chave configurada no Colab
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')

client = Groq(api_key=os.environ.get("GROQ_API_KEY"))

def generate_groq_news(user_name):
    # Chamada para o modelo Llama 3 via Groq
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": "Você é um especialista em marketing bancário. Seja direto e cativante."
            },
            {
                "role": "user",
                "content": f"Crie uma frase curta para {user_name} sobre a importância de investir hoje. (Máximo 100 caracteres)"
            }
        ],
        model="openai/gpt-oss-120b",
    )
    return chat_completion.choices[0].message.content.strip('\"')

# Processando a lista de usuários extraída do CSV
for user in users:
    news_content = generate_groq_news(user['Name'])
    user['news'].append({
        "icon": "",
        "description": news_content
    })
    print(f"Mensagem Groq para {user['Name']}: {news_content}")
```

Load

```python
import json

# Salvando o resultado final em JSON
with open('final_marketing_data.json', 'w', encoding='utf-8') as f:
    json.dump(users, f, ensure_ascii=False, indent=2)

# Visualizando como DataFrame
df_final = pd.DataFrame(users)
print("\n--- Tabela Final de Dados ---")
print(df_final[['Name', 'news']].head())
```
